```python
import pandas as pd                          
import matplotlib.pyplot as plt 
import numpy as np
```

```python
%matplotlib inline                           
%config InlineBackend.figure_format='retina' 
plt.rcParams['figure.figsize'] = 12,6  
```

```python
ees_climate_dir = 'C:/Users/TU_Climate24_Student/Desktop/ees_climate/'  #### This is the home directory we are always working out of

CO2_MaunaLoa = pd.read_csv(ees_climate_dir+'unit2_rexrode/data/greenhouse_gases/CO2_MaunaLoa_TimeSeries.csv',parse_dates=['Date'])
CO2_NOAA = pd.read_csv(ees_climate_dir+'unit2_rexrode/data/greenhouse_gases/CO2_NOAA_TimeSeries.csv',parse_dates=['Date'])
CO2_WDCGG = pd.read_csv(ees_climate_dir+'unit2_rexrode/data/greenhouse_gases/CO2_WDCGG_TimeSeries.csv',parse_dates=['Date'])
```

```python
CO2_MaunaLoa
```

```python
CO2_MaunaLoa.plot(x='Date',
                  y='Mauna Loa CO2 (ppm)',
                  linewidth=2)
```

```python
CO2_NOAA.plot(x='Date',
                  y='NOAA CO2 (ppm)',
                  linewidth=2,
                  color='purple')
```

```python
CO2_WDCGG.plot(x='Date',
                  y='WDCGG CO2 (ppm)',
                  linewidth=2,
                  color='magenta')
```

```python
fig, ax= plt.subplots(figsize=(10,5))

ax.plot(CO2_MaunaLoa['Date'],CO2_MaunaLoa['Mauna Loa CO2 (ppm)'],label='Mauna Loa CO2 (ppm)', color='pink')
ax.plot(CO2_NOAA['Date'],CO2_NOAA['NOAA CO2 (ppm)'],label='NOAA CO2 (ppm)', color='teal')
ax.plot(CO2_WDCGG['Date'],CO2_WDCGG['WDCGG CO2 (ppm)'],label='WDCGG CO2 (ppm)', color='purple')
ax.set_title('CO2 Time Series Comparison')
ax.set_xlabel('Date')
ax.set_ylabel('CO2 Concentration (ppm)')
ax.legend()
ax.grid(True)

plt.show()
```

**Figure 1** CO2 time series comparison from 3 different observatorys.
