```python
import pandas as pd                          
import matplotlib.pyplot as plt 
import numpy as np
import cmocean 

%matplotlib inline                           
%config InlineBackend.figure_format='retina' 
plt.rcParams['figure.figsize'] = 12,6   
```

```python
ees_climate_dir = 'C:/Users/TU_Climate24_Student/Desktop/ees_climate/'
Berkley = pd.read_csv(ees_climate_dir+'unit2_rexrode/data/temp/gmt_Berkeley_Earth.csv')
Era= pd.read_csv(ees_climate_dir+'unit2_rexrode/data/temp/gmt_ERA5.csv')
Gistemp=pd.read_csv(ees_climate_dir+'unit2_rexrode/data/temp/gmt_GISTEMP.csv')
Had=pd.read_csv(ees_climate_dir+'unit2_rexrode/data/temp/gmt_HadCRUT5.csv')
Jra=pd.read_csv(ees_climate_dir+'unit2_rexrode/data/temp/gmt_JRA-55.csv')
Noaa=pd.read_csv(ees_climate_dir+'unit2_rexrode/data/temp/gmt_NOAAGlobalTemp.csv')
```

```python
Noaa
```

```python
fig, ax= plt.subplots(figsize=(10,5))

ax.plot(Berkley['Year'],Berkley['Berkeley Earth (degC)'],label='Berkeley Earth (degC) ', color='pink')
ax.plot(Era['Year'],Era['ERA5 (degC)'],label='ERA5 (degC)', color='teal')
ax.plot(Gistemp['Year'],Gistemp['GISTEMP (degC)'],label='GISTEMP (degC)', color='purple')
ax.plot(Had['Year'],Had['HadCRUT5 (degC)'],label='HadCRUT5 (degC)', color='yellow')
ax.plot(Jra['Year'],Jra['JRA-55 (degC)'],label='JRA-55 (degC)', color='green')
ax.plot(Noaa['Year'],Noaa['NOAAGlobalTemp (degC)'],label='NOAAGlobalTemp (degC)	', color='orange')
ax.set_title('Global Temperature Anomaly Relative to 1850-1900 Comparison')
ax.set_xlabel('Date')
ax.set_ylabel('Global Temperature (degrees Celcius')
ax.legend()
ax.grid(True)

plt.show()
```

**Figure 1** Comparison between 6 recordings of global temperature anomaly relative to 1850-1900 in degrees Celcius. 

```python

from matplotlib import cm
from matplotlib import colormaps as colormap
cool_warm_cmap = colormap["coolwarm"]
cool_warm_cmap
Era_temp_values = Era['ERA5 (degC)'] 
norm = plt.Normalize(min(Era_temp_values), 
                     max(Era_temp_values))

cmap_coolwarm = colormap.get_cmap('coolwarm')
cmap_thermal = colormap.get_cmap(cmocean.cm.thermal)

bar_colors_coolwarm = cmap_coolwarm(norm(Era_temp_values))
bar_colors_thermal = cmap_thermal(norm(Era_temp_values))
fig, ax = plt.subplots(figsize = (10, 5))

ax.bar(Era['Year'],
        Era['ERA5 (degC)'],
        linewidth=2,label='ERA5 global mean surface air temp anomaly',
        color=bar_colors_thermal
        ) 


ax.axhline(0,color='white',linewidth=1)


ax.set_title('Global Ave. Temperature anomaly (deg C) Relative to 1850 - 1900')

ax.set_ylabel('Temperature anomaly (deg C)')
ax.set_xlabel('Year')

ax.set_facecolor("pink")
```

**Figure 2** ERA5 temperature anomaly relative to 1850-1900 in degrees celcius in a bar plot.
